```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n_samples = 13525733
n_fishing = 1671330
n_fishing_labeld = 1628761
n_fishing_non_labeld =  n_fishing - n_fishing_labeld
n_passenger = 1507918

n_merchant = 1939628

timestamp = '2023-05-06'
```

```python
rest = n_samples - n_fishing - n_merchant - n_passenger
```

```python
rest
```

```python
colors = sns.color_palette()
explode = [0.1,0.01,0.01,0.01]
```

```python
# lt.title(f'Distribution of AIS messages on {timestamp}')
plt.pie([n_fishing, n_passenger, n_merchant, rest], labels= ['Fishing', 'Passengers', 'Merchant', 'Other'], colors = colors, autopct = '%0.0f%%', explode = explode, textprops=dict( size=16))
plt.savefig('dist_2023_05_06.png')
```
